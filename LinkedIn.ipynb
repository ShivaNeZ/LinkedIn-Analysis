```python
skill_group_data = data4[data4["skill_group_name"] == "Data Science"]
skill_group_data_2015 = skill_group_data[skill_group_data ["year"]== 2015]["skill_group_penetration_rate"].mean()
skill_group_data_2016 = skill_group_data[skill_group_data ["year"]== 2016]["skill_group_penetration_rate"].mean()
skill_group_data_2017 = skill_group_data[skill_group_data ["year"]== 2017]["skill_group_penetration_rate"].mean()
skill_group_data_2018 = skill_group_data[skill_group_data ["year"]== 2018]["skill_group_penetration_rate"].mean()
skill_group_data_2019 = skill_group_data[skill_group_data ["year"]== 2019]["skill_group_penetration_rate"].mean()
years = [2015, 2016, 2017, 2018, 2019]
average_penetration_rates = [
    skill_group_data_2015,
    skill_group_data_2016,
    skill_group_data_2017,
    skill_group_data_2018,
    skill_group_data_2019
]


plt.figure(figsize=(10, 6))
plt.bar(years, average_penetration_rates, color='navy')
plt.xlabel("Year")
plt.ylabel("Average Skill Group Penetration Rate")
plt.title("Average Data Science Skill Group Penetration Rate (2015-2019)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
country_skill_diversity = data5_3.groupby("country_name")["skill_group_name"].nunique().reset_index()

plt.figure(figsize=(20, 6))
sns.barplot(x="country_name", y="skill_group_name", data=country_skill_diversity)
plt.xlabel("Country")
plt.ylabel("Number of Unique Skill Groups")
plt.title("Skill Diversity by Country")
plt.xticks(rotation=90)
plt.show()

```
